```python
import bcrypt
from pymongo import MongoClient
from config import MONGO_URI, DATABASE_NAME

# connect to your MongoDB
client = MongoClient(MONGO_URI)
db = client[DATABASE_NAME]
users = db["users"]

# Example: update one user
user = users.find_one({"username": "jennifertatalya"})
plain_password = user["password"]  # WARNING: this is plain-text right now

hashed = bcrypt.hashpw(plain_password.encode("utf-8"), bcrypt.gensalt()).decode("utf-8")

users.update_one({"_id": user["_id"]}, {"$set": {"password": hashed}})
print("Password updated and hashed.")
```
